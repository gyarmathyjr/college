# Transcript Analysis

This notebook loads the transcript data from `../data/transcript.json` and displays it as a Pandas DataFrame.

```python
# pandas & numpy
# ..................................................................................... 
import pandas as pd
pd.set_option('display.width', 1000)
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)
pd.set_option('display.max_rows', 250)
pd.set_option('display.max_colwidth', 200)

import numpy as np
np.set_printoptions(precision=2, suppress=True, linewidth=200)
```

```python
import json

# Load the JSON data
with open('../data/transcript.json', 'r') as f:
    data = json.load(f)

# Normalize the nested 'grades' column to flatten it
df = pd.json_normalize(data['course_history'])

# Sort by school_year, requirement_area, and semester
df = df.sort_values(by=['school_year', 'requirement_area', 'course_name', 'semester'])

# Display the DataFrame
df.head()
```
