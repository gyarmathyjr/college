# GPA Analysis

This notebook loads the computed GPA summary from `../data/gpa_summary.json` and displays the results.

```python
import json
import pandas as pd

# Load the GPA summary data
with open('../data/gpa_summary.json', 'r') as f:
    summary = json.load(f)

print(f"Last Updated: {summary['last_updated']}")

# Convert to DataFrame for display
gpa_list = []
for school, data in summary['gpas'].items():
    entry = data.copy()
    entry['school'] = school
    gpa_list.append(entry)
    
df = pd.DataFrame(gpa_list)
# Reorder columns for readability
cols = ['school', 'campus', 'gpa', 'credits_counted', 'semester_grades_counted', 'policy_version']
df = df[cols]

df
```
